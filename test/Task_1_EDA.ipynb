```python
dfs_list = []

for data_type, file_name in list(file_paths.items())[2:3]:
    df = pd.read_csv(file_name, encoding='windows-1251', sep=';')
    print(data_type, len(df), df['Address'].head())
    df['source'] = data_type
    dfs_list.append(df)
    
full_df = pd.concat(dfs_list, axis=0, sort=False, ignore_index=True).reset_index(drop=True)
full_df.columns
```

```python
full_df.describe(include='all')
```

```python
building_pat = re.compile(r'(дом\s+|корпус\s+|строение\s+|владение\s+)')

def not_building(s):
    return not re.search(building_pat, s)
```

```python
full_df['clean_addr'] = full_df['Address'].str.split(',').apply(lambda x: list(filter(not_building, x))[-1].lower().strip())
full_df['clean_addr'].head()
```

```python
print(full_df['clean_addr'].unique().tolist())
```

```python
full_df.groupby('Address')['NumberOfAccessPoints'].sum().sort_values(ascending=False)
```

```python
full_stats_df = full_df.groupby('clean_addr')['NumberOfAccessPoints'].sum().sort_values(ascending=False)
full_stats_df
```
